```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/Loan_default.csv")
```

```python
# Check total null values per column
df.isnull().sum().sort_values(ascending=False)
```

```python
# check if any null values exist at all
print(df.isnull().values.any())
```

```python
# Percentage of missing values per column (helpful for EDA)
print((df.isnull().sum() / len(df)) * 100)
```

Efficiently checking missing data

```python
missing_summary = (
    df.isnull()
      .sum()
      .reset_index()
      .rename(columns={"index": "column", 0: "missing_count"})
)
missing_summary["missing_pct"] = (missing_summary["missing_count"] / len(df)) * 100

print(missing_summary[missing_summary["missing_count"] > 0])
```

Efficient memory usage

```python
print(df.info(memory_usage="deep"))
```

Distribution Plots

```python
for col in df.select_dtypes(include=np.number).columns:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

Correlation Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()
```

Outlier Analysis

```python
for col in df.select_dtypes(include=np.number).columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f"Outlier Analysis - {col}")
    plt.show()
```
